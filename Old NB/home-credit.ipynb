```python
import dask.dataframe as dd
import os
from tqdm import tqdm

# Step 1: Define the directory containing the CSV files
directory = '/kaggle/input/home-credit-credit-risk-model-stability/csv_files/train'

# Step 2: Read all CSV files into separate dask dataframes and store them in a list
dataframes = []
for filename in tqdm(os.listdir(directory)):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        # Specify dtype to handle mixed types
        df = dd.read_csv(filepath, dtype=str)  # Specify types for other columns as needed
        dataframes.append(df)

# Step 3: Perform an outer join on all dataframes based on the 'case_id' column
merged_df = dataframes[0]  # Initialize with the first Dask DataFrame
for df in tqdm(dataframes[1:]):  # Use tqdm to show progress
    # Merge the dataframes, handling duplicate columns by suffixing them
    merged_df = merged_df.merge(df, on='case_id', how='outer', suffixes=('', '_duplicate'))


```

```python
import dask.dataframe as dd
import os
from tqdm import tqdm

# Step 1: Define the directory containing the CSV files
directory = '/kaggle/input/home-credit-credit-risk-model-stability/csv_files/train'

# Step 2: Read all CSV files into separate dask dataframes and store them in a list
dataframes_noDup = []
for filename in tqdm(os.listdir(directory)):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        # Specify dtype to handle mixed types
        df = dd.read_csv(filepath, dtype=str)  # Specify types for other columns as needed
        dataframes_noDup.append(df)

# Step 3: Perform an outer join on all dataframes based on the 'case_id' column
merged_df_noDup = dataframes_noDup[0]  # Initialize with the first Dask DataFrame
for df in tqdm(dataframes_noDup[1:]):  # Use tqdm to show progress
    # Identify columns present in the current dataframe but not in the merged dataframe
    new_columns = [col for col in df.columns if col not in merged_df_noDup.columns]
    # Merge only non-duplicate columns based on 'case_id'
    merged_df_noDup = merged_df_noDup.merge(df[new_columns + ['case_id']], on='case_id', how='inner')

# Drop duplicate 'case_id' column if it exists
#merged_df_noDup = merged_df_noDup.drop_duplicates(subset='case_id')
```
